```python
# this formatter will label the colorbar with the correct target names
formatter = plt.FuncFormatter(lambda i, *args: iris.name[int(i)])

x_name = 'sepal_length'
y_name = 'sepal_width'

plt.scatter(iris[x_name], iris[y_name], c=iris.class_id)

plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.clim(-0.5, 2.5)

plt.xlabel(iris.columns.values[np.where(iris.columns.values == x_name)[0][0]])
plt.ylabel(iris.columns.values[np.where(iris.columns.values == y_name)[0][0]]);
```
