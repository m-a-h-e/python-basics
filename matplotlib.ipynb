```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def classify_array(x, thresholds, classes):
    """
    x          : array to classify
    thresholds : N value sorted array holding the decision boundary thresholds
    classes    : N+1 value sorted array holding the target class values
    
    result     : array of same size as x, filled with related class values
    
    class 0 value is assigned, if x value is smaller than threshold 0
    class 1 value is assigned, if x value is equal or larget than
        threshold 0 and smaller than threshold 1
    etc.
    """
    y = np.full(x.shape, classes[0])
    for n in np.arange(len(thresholds)):
        y[x >= thresholds[n]] = classes[n + 1]
    return y
```

```python
# create some data points ...
num_points = 5000
x = np.random.randn(num_points)
y = x + np.random.randn(num_points)
```

```python
# scatter plot of (x[i],y[i]) points ...
plt.plot(x, y, '.r');
```

```python
# scatter plot of (x[i],y[i]) points ...
plt.scatter(x, y, s=10);
```

```python
# density plot (histogram) of (x[i],y[i]) points ...
density = plt.hist2d(x, y, cmap='coolwarm', bins=30)
```

```python
# mesh plot of given (e.g. density) data ...
plt.pcolormesh(density[1], density[2], density[0].T, cmap='RdYlGn_r');
```

```python
# mesh plot of given (e.g. density) data, overlayed with a scatter plot
plt.pcolormesh(density[1], density[2], density[0].T, cmap='coolwarm');

# sample (xs, ys) points from (x, y) points ...
num_samples = 300
idx = np.random.randint(len(x), size=num_samples)
xs, ys = x[idx], y[idx]

# calculate some zs ...
zs = xs * ys

# classify array into 3 classes using 2 decision boundary thresholds ...
zc = classify_array(zs, [-0.2, 0.2], [0, 1, 2])

# xs, ys and zc are equaly sized numpy vectors
# representing (xs, ys) data points with values zc
plt.scatter(xs, ys, s=10, c=zc, cmap='seismic');
```

```python
# generate mesh grid data ...
x, y = np.meshgrid(np.arange(-2, 2, 0.2), np.arange(-2, 2, 0.2))
z = x*x/3 + y
```

```python
plt.pcolormesh(x, y, z, cmap='coolwarm');
```

```python
plt.contourf(x, y, z, cmap='coolwarm');
```

```python
mn = np.min(z)
mx = np.max(z)
num_levels = 30
levels = np.arange(mn, mx, (mx-mn)/num_levels)

# contour plot (level curve plot) ...
plt.contourf(x, y, z, cmap='coolwarm', levels=levels);
```
